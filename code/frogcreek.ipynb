```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
swe = pd.read_csv('c:/users/marianne/documents/ICB/data/bhv_all.csv')
depth = pd.read_csv('c:/users/marianne/documents/ICB/data/bhv_depth.csv')
```

```python
swe.columns=['date','swe_in','flag']
tmp = [datetime.datetime.strptime(d,'%m/%d/%Y') for d in swe['date']]
swe['dt'] = tmp
swe.swe_in[swe.swe_in == '--']=np.nan
swe.swe_in = swe.swe_in.astype('float')

depth.columns=['date','depth_in','flag']
tmp = [datetime.datetime.strptime(d,'%m/%d/%Y') for d in depth['date']]
depth.depth_in[depth.depth_in == '--']=np.nan
depth.depth_in = depth.depth_in.astype('float')
depth['dt'] = tmp
```

```python
fig,ax= plt.subplots()
ax.plot(swe.dt,swe.swe_in,color = 'red')

fig,ax= plt.subplots()
ax.plot(depth.dt,depth.depth_in,color='black')
```
